```python
country_df = pd.read_excel('Country-Code.xlsx')
country_df
```

```python
# we have COUNTRY CODE in both the zomato.csv and country-code.xlsx files,
#so, we can merge them on the basis of the COUNTRY CODE Column

merged_df = pd.merge(z_df , country_df , on='Country Code' , how='left')
merged_df
```

```python
merged_df.columns
```

```python
# wrt a specific country, how many records are thre?
merged_df.Country.value_counts()
```

```python
# to get the country names
country_names = merged_df.Country.value_counts().index
country_names
```

```python
country_val =  merged_df.Country.value_counts().values   # we're getting the no. of times each country appeared.
country_val
```

```python
## ---  Pie-Chart  ----

#plt.pie( country_val , labels = country_names )    # for all the countries

#plt.pie( country_val[:3] , labels=country_names[:3])   # for the top 3 countries

plt.pie( country_val[:3] , labels=country_names[:3] , autopct = '%1.2f%%')   # for the top 3 countries with PERCENTAGEs

#plt.tight_layout()
#plt.show()
```

```python
merged_df.columns
```

```python
merged_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text'])
```

```python
merged_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size()
```

```python
merged_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size().reset_index()
```

```python
ratings = merged_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size().reset_index().rename( columns = {0:'Rating Count'} )
ratings
```

### Observation -:

1) When rating is b/w 4.5 to 4.9 ---> Excellent

2)                    4.0 to 4.4 ---> Very good

3)                    3.5 to 3.9 ---> Good

4)                    2.5 to 3.4 ---> Average

5)                    1.8 to 2.4 ---> Poor

6)                         0.0   ---> Not Rated

```python
## --- Bar Plot ---

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)  # for the figure size,,,,& when we use this, all figure size changes acc. to the size we gave.

#sns.barplot( x = 'Aggregate rating' , y='Rating Count' , hue='Rating color' , data=ratings )  # we're getting different colors

sns.barplot( x = 'Aggregate rating' , y='Rating Count' , hue='Rating color' , data=ratings , palette = ['grey' , 'red' , 'orange' , 'yellow' , 'green' , 'blue'] )  # to give colors acc to us.

# we're getting a gaussian kind of a curve
```

### Observation -:

1) Not rated count -> very high

2) Max. no. of ratings are b/w 2.5 to 3.4

So, if there's a missing value for rating -> then, we can consider a no. b/w 2.5 to 3.4 as an avg, since, that is the max. no. of rating

```python
## ---  Count Plot  ---

sns.countplot( x='Rating color' , data = ratings , palette = ['grey' , 'red' , 'orange' , 'yellow' , 'green' , 'blue'] )

# here, the 'count' is the frequency, NOT the rating count.
```
